```python
import pandas as pd 
import numpy as np 
```

```python
# df_X_OR = Data Frame Company X - Order Report
df_X_OR = pd.read_excel('Company ShopX - Order Report.xlsx')

# df_X_SKU_M = Data Frame Company X - SKU Master
df_X_SKU_M =pd.read_excel('Company ShopX - Product Weight.xlsx')

# df_X_PZ = Data Frame Company X - Pincode Zones
df_X_PZ = pd.read_excel('Company ShopX - Warehouse&Customer Pin Code and Zone details.xlsx')

# df_CC_I = Data Frame Courier Company - Invoice
df_CC_I = pd.read_excel('Courier Company - Invoice.xlsx')

# df_X_OR = Data  Frame Courier Company - Rates
df_CC_R = pd.read_excel('Courier Company - Rates.xlsx')
```

```python
df_X_OR
Company_X_df = pd.merge(df_X_OR,df_X_SKU_M,on='Product Code')
# converting 'Weight' from float to int
Company_X_df['Weight_calculated'] = Company_X_df['Units Ordered'] * Company_X_df['Product Weight (g)']

DF_1 = Company_X_df.groupby('Order ID')['Weight_calculated'].sum().reset_index()
DF_1['Weight_calculated'] = DF_1['Weight_calculated'].astype(int)
DF_1['Weight (kg)'] = DF_1['Weight_calculated'] / 1000
DF_1['Weight (kg)']= DF_1['Weight (kg)'].round(decimals = 1)
DF_1

# Total weight as per X (KG)
DF_1 = DF_1.rename(columns={'Weight (kg)':'Total weight as per X (KG)'})
DF_1
```

```python
import math

# function to round off Wt Slab as per X
def round_up_to_nearest_half_int(num):
    return math.ceil(num * 2) / 2
```

```python
DF_1['Weight slab as per X (KG)'] = DF_1['Total weight as per X (KG)'].apply(round_up_to_nearest_half_int)
```

```python
def sample_function(row):
    cost=0
    counter = row['Weight_slab_metric']
    if counter == 1 & bool(row['Freight Type']=='Forward charges'):

        cost = row['fwd_fixed']
        status = bool(row['Freight Type']=='Forward charges')
        return cost


    elif counter == 1 &  bool(row['Freight Type']=='Forward and RTO charges'):

        cost = row['rto_fixed']
        return cost

    elif counter > 1 & bool(row['Freight Type']=='Forward charges'):
    
        cost_a=0
        cost_b=0
        final_cost=0
        cost_a = row['fwd_fixed']
        cost_b = row['fwd_additional'] * (counter-1)
        final_cost= cost_a + cost_b
        return final_cost

    else:

        cost_a=0
        cost_b=0
        final_cost=0
        cost_a = row['rto_fixed']
        cost_b = row['rto_additional'] * (counter-1)
        final_cost = cost_a + cost_b
        return final_cost
```

```python
DF_4['Weight_slab_metric'] = DF_4['Weight slab as per X (KG)']/0.5

DF_4['Expected Charge as per X (Rs.)'] =  DF_4.apply(sample_function,axis=1)

DF_4
```
